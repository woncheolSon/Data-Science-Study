배열 연산을 수행할 때는 특이값(outlier)을 제외하거나 적당한 값으로 대체하는 것이 중요하다.

```python
np.random.seed(12345)
```

```python
data = pd.DataFrame(np.random.randn(1000,4))
```

```python
data.tail()
```

```python
data.describe()
```

한 컬럼에서 절대값이 3이 초과하는 값을 찾아내자.

```python
col = data[3]
```

```python
col[np.abs(col>3)]
```

절대값 3을 초과하는 값이 들어있는 모든 로우를 선택하려면 불리언 DataFrame에서 `any` 메서드를 사용하면 된다.

```python
data[(np.abs(data)>3).any(1)]
```

아래 코드로 -3 이나 3을 초과하는 값을 찾을 수도 있다.

유니버셜 함수인 np.sign은 주어진 값의 부호에 따라 1이나 -1이 담긴 배열을 반환한다.

```python
data[np.abs(data)>3] = np.sign(data)*3
```

```python
data.describe()
```
