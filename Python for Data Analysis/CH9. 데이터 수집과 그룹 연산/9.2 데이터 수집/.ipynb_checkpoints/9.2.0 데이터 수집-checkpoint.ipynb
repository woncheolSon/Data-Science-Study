데이터 수집이라고 하면 보통 배열로부터 스칼라 값을 생산해내는 데이터 변환 작업을 떠올리게 된다.<br/>
mean, count, max, min, sum 등과 같은 메서드 외에 직접 고안한 방식과 추가적으로 기존 메서드를 호출하는 방식으로 데이터를 집계할 수 있다.

Series나 DataFrame의 컬럼에 대한 변위치를 계산하는 `quantile`이라는 메서드를 다시 떠올려보자.

```python
df = pd.DataFrame({'key1':['a','a','b','b','a'],
                  'key2':['one','two','one','two','one'],
                  'data1':np.random.randn(5),
                  'data2':np.random.randn(5)})
```

```python
df
```

```python
grouped = df.groupby('key1')
```

```python
grouped['data1'].quantile(0.9)
```

자신만의 데이터 집계 함수를 사용하려면 배열의 `agg`나 `aggregate` 메서드에 해당 함수를 넘기면 된다.

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()
```

```python
grouped.agg(peak_to_peak)
```

```python
grouped.aggregate(peak_to_peak)
```

```python
grouped.describe()
```
