## nba

##### 아래 URL의 NBA 데이터를 크롤링하여 판다스 데이터 프레임으로 나타내세요. 
- http://stats.nba.com/teams/traditional/?sort=GP&dir=-1

```python
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
driver.get('http://stats.nba.com/teams/traditional/?sort=GP&dir=-1')
```

```python
nba_table = driver.find_elements_by_css_selector(".nba-stat-table__overflow")[0]
table_head = list(nba_table.find_element_by_css_selector("thead").text.split(' '))
nba_table_df = pd.DataFrame(columns = table_head)
table_body = nba_table.find_elements_by_css_selector("tbody > tr")
rank = []

for i, team_data in enumerate(table_body): 
    split_team_data = team_data.text.split('\n') 
    rank.append(split_team_data[0])
    data = {
        "TEAM" : split_team_data[1]
            }
    
    team_values = list(split_team_data[2].split(' '))
    for i in range(len(team_values)):
        data[table_head[i+1]]  = team_values[i] 
    nba_table_df.loc[len(nba_table_df)] = data 
    
nba_table_df["rank"] = rank
nba_table_df.set_index("rank")
```

```python
driver.quit()
```

## naver IT/과학

##### 셀레니움을 이용하여 네이버 IT/과학 기사의 10 페이지 까지의 최신 제목 리스트를 크롤링하세요.
- http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105

```python
import time
```

```python
driver = webdriver.Chrome()
driver.get("http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105")
```

```python
def make_url(page=1):
    return "http://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105#&date=%2000:00:00&page="+str(page)

news_titles = []

for i in range(1, 11):
    driver.get(make_url(page = i))
    time.sleep(1)
    news_list = driver.find_elements_by_xpath('//*[@id="main_content"]/div/div[2]/div[1]/div/div[2]/ul/li/div/a')
    for news in news_list:
        news_titles.append(news.text)
news_dict = {
    "News Titles" : news_titles
}
news_df = pd.DataFrame(news_dict)
news_df
```

```python
driver.quit()
```
