##### matplotlib 2
- line plot
- bar chart
- pie chart
- scatter plot
- histogram

```python
y = np.random.randint(5, 10, size=5)
x = range(5)
x, y
```

```python
plt.bar(x, y)
plt.show()
```

```python
# error line
error = np.random.randint(1,5, size=5)
print(x, y, error)
plt.bar(x, y, yerr=error, align='center', alpha=0.8)
plt.show()
```

```python
error = np.random.randint(1,5, size=5)
print(x, y, error)
plt.barh(x, y, xerr=error, align='center', alpha=0.8)
plt.show()
```

```python
y1 = np.random.randint(5, 10, size=5)
y2 = np.random.randint(1, 5, size=5)
print(y1, y2)
x = range(5)
plt.subplots()
plt.bar(x, y1)
plt.bar(x, y2)
plt.show()
```

##### pie chart


```python
y = np.random.randint(1, 10, size=5)
labels = ["A","B","C","D","E"]
colors = ["y","gold","lightskyblue", "lightcoral", "r"]
plt.pie(y, labels=labels, colors=colors, autopct="%1.1f%%")
plt.show()
```

##### scatter plot
- 균등분포, 정규분포

```python
# 균등분포
x1 = np.random.rand(500)
y1 = np.random.rand(500)
plt.scatter(x1, y1)
plt.show()
```

```python
# 정규분포
x1 = np.random.randn(500)
y1 = np.random.randn(500)
plt.scatter(x1, y1)
plt.show()
```

```python
# 균등분포
x1 = np.random.randint(1,100, size=500)
y1 = np.random.randint(1,100, size=500)
plt.scatter(x1, y1)
plt.show()
```

```python
x1 = np.random.randint(1,10,size=20)
y1 = np.random.randint(1,10,size=20)
colors = np.random.rand(20)
area = np.pi * np.arange(10,50,2) ** 2
plt.scatter(x1, y1, s=area, c=colors, alpha=0.6)
plt.show()
```

##### histogram

```python
x1 = np.random.rand(1000)
plt.hist(x1, bins=20)
plt.show()
```

```python
x1 = np.random.randn(1000)
plt.hist(x1, bins=20)
plt.show()
```
