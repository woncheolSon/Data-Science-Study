```python
def sinplot():
    x = np.linspace(0, 10, 101) # 0 ~ 10데이터를 101등분
    for i in range(6):
        plt.plot(x, np.sin(i + x))
sinplot()
```

```python
# histogram
sns.set()
sns.rugplot(iris.petal_length.values)
```

```python
sns.kdeplot(iris.petal_length.values)
```

```python
sns.distplot(iris.petal_length.values, kde=False)
```

```python
sns.distplot(iris.petal_length.values, rug=True)
```

##### joint_plot

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
sns.jointplot(x="petal_length", y="petal_width", data=iris)
```

##### pairplot

```python
sns.pairplot(iris, hue="species")
plt.show()
```

##### lmplot

```python
sns.lmplot(x="petal_length", y="petal_width", hue="species", data=iris, markers=["+","o","*"])
```

##### heatmap

```python
tips = sns.load_dataset("tips")
tips.tail()
```

```python
# pivot table index:time, column:sex, value:tip
```

```python
tips_df = tips.pivot_table("tip", "time","sex", aggfunc=np.mean)
tips_df
```

```python

sns.heatmap(tips_df, cmap=sns.light_palette("purple", as_cmap=True), annot=True)
```

```python
flights = sns.load_dataset("flights")
flights.tail()
```

```python
flights_df = flights.pivot("month", "year", "passengers")
flights_df
```

```python
sns.heatmap(flights_df, cmap="YlGnBu", annot=True, fmt="d")
```
