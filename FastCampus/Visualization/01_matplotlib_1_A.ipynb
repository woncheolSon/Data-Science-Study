```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Basic Line Plot
data = np.random.randint(10, size=10)
print(data)
```

```python
plt.plot(data)
plt.show()
```

```python
# x데이터 추가 - x축 데이터 단위
y = np.random.randint(10, size=10)
x = range(0,100,10)
print(list(x), list(y))
plt.plot(x, y)
plt.show()
```

##### 스타일

```python
print(data)
plt.plot(data, 'r*:')
plt.show()
```

```python
print(data)
plt.plot(data, c="b", lw=5, ls="-.", marker="*", ms=10, mec="r", mew=1, mfc="g")
plt.show()
```

```python
# limit 설정
plt.xlim(-5, 20)
plt.ylim(-10, 20)
plt.plot(data)
plt.show()
```

```python
# xtick, ytick 설정
plt.plot(data)
plt.xticks(range(0,20,2))
plt.yticks([1,2,3,4,5,6])
plt.show()
```

```python
# 그리드 추가
plt.grid(True)
plt.plot(data)
plt.show()
```

```python
# 멀티 라인 그리기
data1 = data
data2 = np.random.randint(10, size=10)
data3 = np.random.randint(10, size=10)
x = range(10)
print(data1, data2, data3, x)

plt.plot(x, data1, 'ro-', x, data2, 'b^-.', x, data3, 'g*--')
plt.show()
```

```python
plt.plot(x, data1)
plt.plot(x, data2)
plt.plot(x, data3)
plt.show()
```

```python
# 범례 - legend
data1 = np.random.randint(10, size=10)
data2 = np.random.randint(10, size=10)
plt.plot(x, data1, label="data1")
plt.plot(x, data2, label="data2")
plt.legend(loc=0)
plt.show()
```

```python
# 타이틀 및 x축 y축 라벨링
plt.plot(data)
plt.title("Title")
plt.ylabel("ylabel")
plt.xlabel("xlabel")
plt.show()
```

```python
plt.plot(data)
plt.title("타이틀")
plt.ylabel("y축")
plt.xlabel("x축")
plt.show()
```

##### 한글폰트 적용

    - matplotlib 폰트 설정파일을 찾아서 설정을 바꿔 놓는것
    - 폰트의 경로를 설정해서 사용하는 방법

```python
# 방법 1 - 설정파일을 수정
```

```python
# 현재 폰트를 확인
mpl.rcParams["font.family"]
```

```python
# 설정파일의 위치
mpl.matplotlib_fname()
```

```python
# 설정파일에서 아래의 코드를 삽입
# font.family : AppleGothic
```

```python
# 방법 2 - 폰트의 경로를 지정
# 폰트의 위치
# 윈도우 - C:\\Windows\Fonts\malgunbd.ttf
# 맥 - /Library/Fonts/AppleGothic.ttf
import matplotlib.font_manager as fm
font_location = "/Library/Fonts/AppleGothic.ttf"
font_name = fm.FontProperties(fname=font_location).get_name()
print(font_name)
mpl.rc('font', family=font_name)
```

```python
plt.subplot(3,1,1) # 3*1 행렬 1번째에 그린다.
plt.plot(x, data1)

plt.subplot(3,1,2)
plt.plot(x, data2)

plt.subplot(3,1,3)
plt.plot(x, data3)

plt.show()
```

```python
# , 생략이 가능
plt.subplot(221) 
plt.plot(x, data)

plt.subplot(222)
plt.plot(x, data1)

plt.subplot(223)
plt.plot(x, data2)

plt.subplot(224)
plt.plot(x, data3)

plt.show()
```
