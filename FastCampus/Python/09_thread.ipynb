##### Thread
- 프로그램을 실행하면서 시간이 너무 걸린다면 두가지 문제가 있을수 있습니다.
- I/O 바운드
    - 대부분의 경우의 문제, 데이터를 넣거나 뺄때 걸리는 시간, 네트워크 트래픽에 소요되는 시간등
    - 비동기 처리를 이용해서 해결이 가능합니다.
- CPU 바운드
    - 계산량이 굉장히 많은 경우에 발생하게 됩니다.
    - 알고리즘이나 하드웨어 성능으로 해결이 가능합니다.
    - 가격 : CPU-ALU(연산처리장치) >>>>> GPU-ALU(연산처리장치)
- 동기 sync
    - 한가지 작업이 끝나야 다음 작업을 수행
- 비동기 async
    - 작업이 독립적으로 수행 됩니다. 작업이 끝나는 시간을 알수 없습니다.

```python
import threading

ls1, ls2 = [], []
```

```python
def th_func1(number=1000):
    for idx in range(number):
        ls1.append(idx)
        if idx % 100 == 0:
            print("ls1 :", idx)
            
def th_func2(number=1000):
    for idx in range(number):
        ls2.append(idx)
        if idx % 100 == 0:
            print("ls2 :", idx)
                        
```

```python
th1 = threading.Thread(target=th_func1, args=(12000,))
th2 = threading.Thread(target=th_func2, args=(10000,))
th1.start()
th2.start()
```
