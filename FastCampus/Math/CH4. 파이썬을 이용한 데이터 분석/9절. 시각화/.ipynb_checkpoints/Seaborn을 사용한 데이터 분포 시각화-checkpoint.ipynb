# Seaborn을 사용한 데이터 분포 시각화

### 스타일

```python
# seaborn을 임포트하면 색상 등을 matplotlib에서 제공하는 기본 스타일이 아닌 seaborn에서 
# 지정한 기본 스타일로 바꾼다.
# 따라서 동일한 matplotlib 명령을 수행해도 seaborn을 임포트 한 것과,
# 하지 않은 플롯의 모양이 다르다.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

- `set`: 색상, 틱 스타일 등 전반적인 플롯 스타일을 seaborn 기본 스타일로 바꾼다.
- `set_stlye`: 틱 스타일만 바꿀 수 있다.
    - `darkgrid`,`whitegrid`,`dark`,`white`,`ticks`
- `set_color_codes`: 기본 색상을 가리키는 문자열을 바꾼다.

```python
sns.set()
sns.set_style('whitegrid')
sns.set_color_codes()
```

### 색상 팔렛트

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
sns.palplot(sns.color_palette("Blues"))
```

```python
sns.palplot(sns.dark_palette("muted purple", input="xkcd"))
```

### 1차원 분포 플롯

- 1차원 데이터는 실수 값이면, 히스토그램과 같은 실수 분포 플롯으로 나타내고
              카테고리 값이면, 카운트 플롯으로 나타낸다.

```python
iris = pd.read_csv("../seaborn-data/iris.csv")    # 붓꽃 데이터
titanic = pd.read_csv("../seaborn-data/titanic.csv")    # 타이타닉호 데이터
tips = pd.read_csv("../seaborn-data/tips.csv")    # 팁 데이터
flights = pd.read_csv("../seaborn-data/flights.csv")    # 여객운송 데이터
```

### 1차원 실수 분포 플롯

- 실수 분포 플롯은 자료의 분포를 묘사하기 위한 것으로 matplotlib의 단순한 히스토그램과 달리 커널 밀도(kernel density) 및 러그(rug) 표시 기능 다차원 복합 분포 기능 등을 제공한다.
- `rugplot`,`kdeplot`,`distplot`

```python
# 러그(rug) 플롯은 데이터 위치를 x축 위에 작은 선분(rug)으로 나타내어 
# 실제 데이터들의 위치를 보여준다.
```

```python
x = iris.petal_length.values
```

```python
sns.rugplot(x)
# plt.show()
```

```python
# 커널 밀도(kernel density)는 커널이라는 함수를 겹치는 방법으로 히스토그램보다
# 부드러운 형태의 분포 곡선을 보여주는 방법이다.
```

```python
sns.kdeplot(x)
```

```python
sns.distplot(x, kde=True, rug=True)
```

### 카운트 플롯

- `countplot`: 각 카테고리 값별로 데이터가 얼마나 있는지 표시할 수 있다.
    - 데이터프레임에만 사용할 수 있다.
    - 사용 방법: `countplot(x="column_name", data=dataframe)`

```python
sns.countplot(x="class", data=titanic)
```

```python
sns.countplot(x="day", data=tips)
```

### 다차원 데이터
- 데이터 변수가 여러개인 다차원 데이터는 데이터의 종류에 따라 다음과 같은 경우가 있을 수 있다.
    - 데이터가 모두 실수 값인 경우
    - 데이터가 모두 카테고리 값인 경우
    - 데이터가 모두 실수 값과 카테고리 값이 섞여 있는 경우

### 2차원 실수형 데이터

- 만약 데이터가 2차원이고 모두 연속적인 실수값이라면 스캐터 플롯(scatter plot)을 사용하면 된다. 
- `jointplot` 명령은 스캐터 플롯뿐 아니라 차트의 가장자리(margin)에 각 변수의 히스토그램도 그린다.
    - 데이터프레임에만 사용할 수 있다.
    - 사용방법: `jointplot(x="x_name", y="y_name", data=dataframe, kind='scatter')`
        - `x`: x 변수가 될 데이터프레임의 열 이름 문자열
        - `y`: y 변수가 될 데이터프레임의 열 이름 문자열
        - `data`: 대상이 되는 데이터프레임
        - `kind`: 차트의 종류를 지정

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
```

- `kind = kde`: 커널 밀도 히스토그램을 그린다.

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
```

### 다차원 실수형 데이터

- 만약 3차원 이상의 데이터라면 seaborn 패키지의 `pairplot` 명령을 사용한다. 
- `pairplot` 은 데이터프레임을 인수로 받아 그리드(grid) 형태로 각 데이터 열의 조합에 대해 스캐터 플롯을 그린다. 
- 같은 데이터가 만나는 대각선 영역에는 해당 데이터의 히스토그램을 그린다.

```python
sns.pairplot(iris)
```

- 만약 카테고리형 데이터가 섞여 있는 경우에는 `hue` 인수에 카테고리 변수 이름을 지정하여 카테고리 값에 따라 색상을 다르게 할 수 있다.

```python
iris
```

```python
sns.pairplot(iris, hue="species", markers=["o","s","D"])
```

### 2차원 복합 데이터

- 만약 데이터가 2차원이고 실수 값, 카테고리 값이 섞여 있다면 기존의 플롯 이외에도 아래와 같은 분포 플롯들을 이용할 수 있다.
    - `barplot`,`boxplot`,`pointplot`,`violinplot`,`stripplot`,`swarmplot`

- `barplot`: 카테고리 값에 따른 실수 값의 평균과 편차를 표시하는 기본적인 바차트
    - 평균은 막대의 높이
    - 편차는 에러바(error bar)로 표시한다.

```python
tips.tail()
```

```python
sns.barplot(data=tips, x="day", y="total_bill")
```

- `boxplot`
    - 박스는 실수 값 분포에서 1사분위수(Q1)와 3사분위수(Q3)를 뜻하고 이 3사분위수와 1사분위수의 차이 (Q3-Q1)를 IQR(interquartile range)라고 한다. 
    - 박스 내부의 가로선은 중앙값을 나타낸다. 박스 외부의 세로선은 1사분위 수보다 1.5 x IQR 만큼 낮은 값과 3사분위 수보다 1.5 x IQR 만큼 높은 값의 구간을 기준으로 그 구간의 내부에 있는 가장 큰 데이터와 가장 작은 데이터를 잇는 선분이다.
    - 그 바깥의 점은 아웃라이어(outiler)라고 부르는데 일일히 점으로 표시한다.

```python
sns.boxplot(data=tips, x="sex", y="total_bill")
```

- `boxplot` 이 중앙값, 표준편차 등 분포의 간략한 특성한 보여주는데 반해,
- `violinplot`,`stripplot`,`swarmplot` 등은 카테고리값에 따른 각 분포의 실제 데이터나 전체 형상을 보여준다는 장점이 있다

```python
# violinplot은 세로방향으로 커널 밀도 히스토그램을 그려주는데
# 왼쪽과 오른쪽이 대칭이 되도록 하여 바이올린처럼 보인다.
sns.violinplot(data=tips, x="sex", y="total_bill")
```

```python
# stripplot은 마치 스캐터 플롯처럼 모든 데이터를 점으로 그려준다.
# jitter=True를 설정하면 가로축 상의 위치를 무작위로 바꾸어서 데이터의 수가 많을 경우에
# 겹치지 않도록 한다.

# np.random.seed(0)
sns.stripplot(data=tips, x="day", y="total_bill",jitter=True)
```

```python
# swarmplot은 stripplot 과 비슷하지만 데이터를 나타내는 점이 겹치지 않도록 옆으로 이동한다.
```

```python
sns.swarmplot(data=tips, x="day", y="total_bill")
```

### 다차원 복합 데이터

- 지금까지 소개한 대부분의 명령어에는 2차원 이상의 고차원 데이터에 대해서도 분석할 수 있는 기능이 포함되어 있다.
- 예를들어 `barplot`,`violinplot`,`boxplot` 등 에서는 두가지 카테고리 값에 의한 실수 값의 변화를 보기 위한 `hue` 인수를 제공한다.
- `hue`인수에 카테고리 값을 가지는 변수의 이름을 지정하면 카테고리 값에 따라 다르게 시각화 된다.
- `hue` 값이 시각화되는 방법은 플롯의 종류에 따라 다르다.

```python
# hue 없이
sns.barplot(data=tips, x="day",y="total_bill")
```

```python
# hue 포함
sns.barplot(data=tips, x="day",y="total_bill", hue="sex")
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
# hue 없이
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True)
```

```python
# hue 포함
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", hue="sex", data=tips, jitter=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

- `stripplot`,`violinplot`,`swarmplot` 등 에서는 `split` 또는 `dodge` 옵션으로 시각화 방법을 변경할 수도 있다.
- `split`: 두개의 그래프에 표시되던 것을 하나의 그래프에 반반씩 나타냄
- `dodge`: 하나의 그래프에 표시되던 것을 두개의 그래프로 각각 나타냄, split의 반대

```python
# split 없이
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex")
```

```python
# split 포함
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
```

```python
# dodge 없이
sns.stripplot(x="day", y="total_bill", hue="sex", data=tips, jitter=True)
```

```python
# dodge 포함
sns.stripplot(x="day", y="total_bill", hue="sex", data=tips, jitter=True, dodge=True)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips, dodge=True)
```

- `heatmap`: 두 개의 카테고리 값에 의한 실수 값 변화를 볼 수 있다.

```python
flights.tail()
```

```python
fllghts_passengers = flights.pivot("month", "year")
fllghts_passengers
```

```python
fllghts_passengers = flights.pivot("month", "year","passengers")
fllghts_passengers
```

```python
fllghts_passengers = flights.pivot("month", "year","passengers")
sns.heatmap(fllghts_passengers,annot=True,fmt="d",linewidths=1)
```

- `factorplot`: 색상(hue)과 행(row) 등을 동시에 사용하여 3개 이상의 카테고리 값에 의한 분포 변화를 보여준다.

```python
titanic.tail()
```

```python
data = titanic[titanic.survived.notnull()]
data.tail()
```

```python
sns.factorplot(data=data, x="age",y="sex",hue="survived",row="class",
               kind="violin",split=True, size=2, aspect=4)
```
