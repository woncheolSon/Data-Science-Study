```python
%matplotlib nbagg
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
battles = pd.read_csv("game-of-thrones/battles.csv", sep=",")
deaths = pd.read_csv("game-of-thrones/character-deaths.csv", sep=",")
```

```python
book_nums_to_death_count = deaths["Book of Death"].value_counts().sort_index()
```

```python
book_nums_to_death_count
```

```python
ax1 = book_nums_to_death_count.plot(color="k", marker="o", linestyle="--")
```

```python
ax1.set_xticks(np.arange(1,6))
```

```python
ax1.set_xlim([0,6])
```

```python
ax1.set_ylim([0,120])
```

```python
battles.head()
```

```python
battles = battles.set_index(["name"])
```

```python
battles.head()
```

```python
large_battles_mask = battles["attacker_size"]+battles["defender_size"]>10000
```

```python
large_battles = battles.loc[large_battles_mask, ["attacker_size","defender_size"]]
```

```python
large_battles.head()
```

```python
ax2 = large_battles.plot(kind="barh",stacked=True, fontsize=8)
```

```python
large_battles["attacker_pcts"] = \
    large_battles["attacker_size"] / (large_battles["attacker_size"]+large_battles["defender_size"])
```

```python
large_battles["defender_pcts"] = \
    large_battles["defender_size"] / (large_battles["attacker_size"]+large_battles["defender_size"])
```

```python
ax3 = large_battles[["attacker_pcts", "defender_pcts"]].plot(kind="barh", stacked=True, fontsize=8)
```

```python
battles.head()
```

```python
col_names = battles.columns[4:12]
```

```python
col_names
```

```python
houes_names = battles[col_names].fillna("None").values
```

```python
houes_names[:5]
```

```python
houes_names = np.unique(houes_names)
```

```python
houes_names
```

```python
houes_names = houes_names[houes_names!="None"]
```

```python
houes_names
```

```python
house_to_battle_counts = pd.Series(0, index=houes_names)
```

```python
house_to_battle_counts
```

```python
for col in col_names:
    house_to_battle_counts = house_to_battle_counts.add(battles[col].value_counts(), fill_value=0)
```

```python
house_to_battle_counts
```

```python
ax4 = house_to_battle_counts.hist(bins=10)
```
