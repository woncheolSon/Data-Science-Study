```python
df2 = pd.DataFrame({'a':range(7), 'b':range(7,0,-1),
                    'c':['one','one','one','two','two','two','two'],
                    'd':list("0120123")})
```

```python
df2
```

```python
df3 = df2.set_index(['c','d']) # 기존의 열을 index 로 변환
```

```python
df3
```

```python
df2.set_index(['c','d'], drop=False) #drop 는 기존열을 유지함과 동시에 index로 활용할수있게함
```

```python
df3
```

```python
df3.reset_index() # 기존 index를 column으로 전환하고 index는 정수열로 변환함
```

```python
df4 = pd.DataFrame(np.arange(6).reshape((2,3)),
                   index=['Seoul', 'Busan'],
                   columns=['one','two','three'])
df4.index.name = "city"
df4.columns.name = 'number'
```

```python
df4
```

```python
df5 = df4.stack() # 기존의 열을 최하위 index로 변환
```

```python
df5
```

```python
df5.unstack()
```

```python
df5.unstack(level=0)
```

```python
df5.unstack(level="city")
```

```python
s1 = pd.Series([0,1,2,3], index=['a','b','c','d'])
s2 = pd.Series([4,5,6], index=['c','d','e'])
s3 = pd.concat([s1,s2], keys=['one','two'])
```
