표시용 DataFrame 안에 있는 컬럼에 접두어(prefix)를 추가한 후 다른 데이터와 병합하고 싶을 때가 있는데,<br/>
이런 경우 get_dummies 함수의 prefix 인자를 사용하면 된다.

```python
dummies = pd.get_dummies(df['key'],prefix='key')
```

```python
dummies
```

```python
df_with_dummy = df[['data1']].join(dummies)
```

```python
df_with_dummy
```

DataFrame의 한 컬럼이 여러 카테고리에 속한다면 일이 조금 복잡해지는데,<br/>
앞에서 살펴봤던 MovieLens의 영화 데이터를 다시 살펴보자.

```python
mnames = ['movie_id','title','genres']
```

```python
movies = pd.read_table('../data/movies.dat',sep='::',header=None, names=mnames)
```

```python
movies[:10]
```

각 장르마다 표시용 값을 추가하려면 약간의 수고를 해야 하는데, 먼저 데이터 묶음에서 유일한 장르 목록을 추출해야 한다 (`set.union` 트릭을 사용하면 된다.)

```python
genre_iter = (set(x.split(':')) for x in movies.genres)
```

```python
genres = sorted(set.union(*genre_iter))
```

```python
genres
```

이제 표시용 DataFrame을 생성하기 위해 0으로 초기화된 DataFrame을 생성하자.

```python
dummies = pd.DataFrame(np.zeros((len(movies),len(genres))), columns=genres)
```

```python
dummies.tail()
```

```python
for i, gen in enumerate(movies.genres):
    dummies.ix[i, gen.split(':')]=1
```

```python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
```

```python
movies_windic.ix[0]
```

```python
movies_windic
```

`get_dummies` 나 `cut` 같은 이산 함수를 잘 조합하면 통계 어플리케이션에서 유용하게 사용할 수 있다.

```python
values = np.random.rand(10)
```

```python
values
```

```python
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]
```

```python
pd.get_dummies(pd.cut(values,bins))
```
