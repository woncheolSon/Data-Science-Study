집계는 그룹연산의 한 종류일 뿐이다. 1차원 배열을 스칼라 값으로 줄여주는 함수를 적용하는, 일반적인 데이터 변형의 한 가지 특수한 경우이다.<br/>
이 절에서는 다양한 그룹 연산을 수행할 수 있는 `transform`과 `apply` 메서드에 대해 알아본다.

```python
from seaborn import load_dataset
```

```python
df = pd.DataFrame({'key1':['a','a','b','b','a'],
                  'key2':['one','two','one','two','one'],
                  'data1':np.random.randn(5),
                  'data2':np.random.randn(5)})
```

```python
df
```

```python
k1_means = df.groupby('key1').mean()
k1_means
```

```python
k1_means = df.groupby('key1').mean().add_prefix('mean_')
k1_means
```

```python
pd.merge(df, k1_means, left_on='key1', right_index=True)
```

```python
people = pd.DataFrame(np.random.randn(5,5),
                     columns=['a','b','c','d','e'],
                     index=['joe','steve','wes','jim','travis'])
```

```python
people
```

```python
key = ['one','two','one','two','one'] # index
```

```python
people.groupby(key).mean()
```

```python
people.groupby(key).transform('mean')
```

```python
people.groupby(key).transform(np.mean)
```

각 그룹에서 평균값을 빼고 싶다면 함수를 하나 만들어 `transform`메서드에 넘기면 된다.

```python
def demean(arr):
    return arr - arr.mean()
```

```python
demeaned = people.groupby(key).transform(demean)
```

```python
demeaned
```

```python
demeaned.groupby(key).mean()
```
